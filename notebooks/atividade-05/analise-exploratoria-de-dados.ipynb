# Analisando a lista de esportes

```python
df["Sport"].unique()
```

## Quantidade de atletas em cada esporte

```python
df["Sport"].value_counts()
```

```python
data = df["Sport"].value_counts().reset_index()
data = data.nlargest(5, 'Sport')
```

```python
sns.set_style("whitegrid")
plt.figure(figsize=(5, 2))
ax = sns.barplot(data=data, x='index', y='Sport', color='tomato')
ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type='center',
    color='white'
)
plt.xlabel('Sports')
plt.ylabel('N° de Atletas')
plt.title('Tipos de esportes')
ax.set(frame_on=False)
plt.show()
```

# Salario dos atletas em cada esporte

```python
sns.set_style('whitegrid')
plt.figure(figsize=(8,8))
axis=sns.barplot(
            y=df.Sport,
            x=df['earnings ($ million)'],
            color='tomato',
            errorbar=None    
                )

axis.set(frame_on=False)
axis.axvline(0, color='k', clip_on=False)
plt.title('Lista dos Esportes com Maiores Salarios')
plt.xlabel('Milhões $')
plt.ylabel('Tipos de Esportes')
plt.show()
```

# Historico de salarios com o passar dos anos

```python
sns.set_style('whitegrid')
plt.figure(figsize=(20,10))
axis=sns.barplot(data=df,
            x='Year',
            y=df['earnings ($ million)'],
            color='brown',
            errorbar=None   
                )

axis.set(frame_on=False)
axis.axhline(0, color='k', clip_on=False)
plt.show()
```
